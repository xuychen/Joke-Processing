```python
from sklearn.manifold import TSNE

import pandas as pd
import matplotlib.pyplot as plt
from ggplot import ggplot, aes, geom_point, ggtitle
%matplotlib inline
```

```python
# get data
data = pd.read_csv("data-outliers_removed.csv")
data = data.drop(data.columns[63:], axis=1)
```

```python
# t-SNE visualization 
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=1000)
result = pd.DataFrame(tsne.fit_transform(data), columns=['x-tsne', 'y-tsne'])
result = result.join(data)

for i in range(2, result.shape[1]):
    chart = ggplot(result, aes(x='x-tsne', y='y-tsne', color=result.columns[i]))\
            + geom_point(size=70, alpha=1) \
            + ggtitle("tSNE dimensions colored by digit")

    print chart
```
